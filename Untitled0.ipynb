````python
import matplotlib.pyplot as plt
import seaborn as sns
#Countplot bassed on gender

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender')

plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
````

````python
#Bar chart for total spending by gender


gender_sales = df.groupby('Gender')['Amount'].sum()
plt.figure(figsize=(6,4))
gender_sales.plot(kind='bar')

plt.title('Total Spending by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.show()
````

````python
# Create Age Group column from Age

bins = [0, 17, 25, 35, 45, 55, 100]
labels = ['0-17', '18-25', '26-35', '36-45', '46-55', '55+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
````

````python
# Count Plot for Age Group

plt.figure(figsize=(8,4))
sns.countplot(data=df, x='Age Group')
plt.title('Count of Customers by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
````

````python
# Total Spending by Age Group

plt.figure(figsize=(8,4))
df.groupby('Age Group')['Amount'].sum().plot(kind='bar')
plt.title('Total Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Spending')
plt.show()
````
